### 
- Phân loại các quốc gia thành các nhóm dựa trên số lượng ca nhiễm 


```python
def group_ca_nhiem(value):
    if value <= 100:
        return '0-100'
    elif value <= 1000:
        return '101-1000'
    elif value <= 50000:
        return '1001-50000'
    elif value <= 100000:
        return '50001-100000'
    elif value <= 150000:
        return '100001-150000'
    elif value <= 200000:
        return '150001-200000'
    elif value <= 250000:
        return '200001-250000'
    elif value <= 300000:
        return '250001-300000'
    elif value <= 350000:
        return '300001-350000'
    else:
        return '350001-400000'
df['group'] = df['Confirmed'].apply(group_ca_nhiem)
```

bảng cross-tabulation. 

```python
crosstab_confirmed = pd.crosstab(index=df['Country/Region'], columns=df['Year'], values=df['Confirmed'], aggfunc='sum')
crosstab_deaths = pd.crosstab(index=df['Country/Region'], columns=df['Year'], values=df['Deaths'], aggfunc='sum')
crosstab_recovered = pd.crosstab(index=df['Country/Region'], columns=df['Year'], values=df['Recovered'], aggfunc='sum')
crosstab_df = pd.concat([crosstab_confirmed, crosstab_deaths, crosstab_recovered], axis=1, keys=['Confirmed', 'Deaths', 'Recovered'])
crosstab_df
```
