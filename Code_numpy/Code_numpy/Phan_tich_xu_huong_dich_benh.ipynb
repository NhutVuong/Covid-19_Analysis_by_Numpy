```python
import numpy as np
filepath = "C:\\fpoly\\Numpy\\Assignment\\"
file_full_name = "Covid-19-Final.csv"
filename_full = filepath + file_full_name

file_name_Country = "Covid-19-Country.csv"
filename_country = filepath + file_name_Country
```

```python
myDelimiter = ','
maxline = 10
np.set_printoptions(suppress = True, precision = 2, linewidth = 200, formatter={'float': '{:20,.1f}'.format})
```

```python
country_data = np.genfromtxt(filename_country,
                     delimiter = myDelimiter,
                     dtype = np.str_,
                   skip_header = 1)
#print(country_data)
```

```python
indeces_raw = [0,8,9,10,5,6,7,11,12]
raw_data = np.genfromtxt(filename_full,
                        usecols = indeces_raw,
                        delimiter = myDelimiter,
                        skip_header = True)
print(raw_data)
print(raw_data.shape)
SNo_cIdx = 0
MM_cIdx = 1
DD_cIdx = 2
YYYY_cIdx = 3
cofirmed_cIdx = 4
deaths_cIdx = 5
recoverds_cIdx = 6
stateid_cIdex = 7
countryid_cIdex = 8
```

- Mức 1: Thông tin Covid-19 theo từng năm

```python
print(raw_data)
print(raw_data.shape)
```

```python
unique_yyyy = np.unique(raw_data[:,YYYY_cIdx])
print(unique_yyyy)
```

```python
groupbyYY_column_names = ['STT','SUm_of_Cofirmed','Sum_of_Deaths','Sum_of_Recovered','YYYY']
NumofRows = len(unique_yyyy)
NumofColumns = len(groupbyYY_column_names)

data_groupby_YYYY = np.full((NumofRows,NumofColumns),0.)
grp_STT_idx = 0
grp_Sum_Confirmed_idx = 1
grp_Sum_Deaths_idx = 2
grp_Sum_Recovered_idx = 3
grp_YYYY_idx = 4

print(groupbyYY_column_names)
print(data_groupby_YYYY)
print(data_groupby_YYYY.shape)
```

```python
for rowI in range(0,NumofRows):
    selectedYYYY = unique_yyyy[rowI]
    
    rIdx_data_byYYYY = np.where(raw_data[:,YYYY_cIdx] == selectedYYYY )
    data_byYYYY = raw_data[rIdx_data_byYYYY[0]]
    print(data_byYYYY.shape)
    
    data_groupby_YYYY[rowI,grp_STT_idx] = rowI + 1
    
    data_groupby_YYYY[rowI,grp_Sum_Confirmed_idx] = data_byYYYY[:,cofirmed_cIdx].sum()
    data_groupby_YYYY[rowI,grp_Sum_Deaths_idx] = data_byYYYY[:,deaths_cIdx].sum()
    data_groupby_YYYY[rowI,grp_Sum_Recovered_idx] = data_byYYYY[:,recoverds_cIdx].sum()
    
    data_groupby_YYYY[rowI,grp_YYYY_idx] = selectedYYYY
    print(data_groupby_YYYY[rowI])
    
header_str = ",".join(groupbyYY_column_names)
np.savetxt(filepath+'Total_Covid-19_by_Year.csv',data_groupby_YYYY,delimiter = myDelimiter,header = header_str,fmt = "%d,%d,%d,%d,%d")   
```

### 
- Mức 2: Thông tin Covid-19 của các tháng theo năm

```python
unique_yyyy_mm = np.unique(raw_data[:, [YYYY_cIdx, MM_cIdx]], axis=0)
sorted_indices = np.lexsort((unique_yyyy_mm[:, 1], unique_yyyy_mm[:, 0]))
unique_yyyy_mm_sorted = unique_yyyy_mm[sorted_indices]
print(unique_yyyy_mm_sorted)
```

```python
for rowI in range(0,NumofRows_yyyy_mm):
    selectedYYYY,selectedMM = unique_yyyy_mm_sorted[rowI]
    rIdx_data_byYYYYMM = np.where((raw_data[:, YYYY_cIdx] == selectedYYYY) &
                                  (raw_data[:, MM_cIdx] == selectedMM))
    data_byYYYYMM = raw_data[rIdx_data_byYYYYMM[0]]
    print(data_byYYYYMM.shape)
    
    data_groupby_YYYYMM[rowI, grp_STT_idx] = rowI + 1
    
    data_groupby_YYYYMM[rowI, grp_Sum_Confirmed_idx] = data_byYYYYMM[:, cofirmed_cIdx].sum()
    data_groupby_YYYYMM[rowI, grp_Sum_Deaths_idx] = data_byYYYYMM[:, deaths_cIdx].sum()
    data_groupby_YYYYMM[rowI, grp_Sum_Recovered_idx] = data_byYYYYMM[:, recoverds_cIdx].sum()
    data_groupby_YYYYMM[rowI, grp_YYYY_idx] = selectedYYYY
    data_groupby_YYYYMM[rowI, grp_MM_idx] = selectedMM
    print(data_groupby_YYYYMM[rowI])
```

```python
header_str_2 = ",".join(groupbyYYMM_column_names)
np.savetxt(filepath+'Total_Covid-19_by_YearMonth.csv',data_groupby_YYYYMM,delimiter = myDelimiter,header = header_str_2,fmt = "%d,%d,%d,%d,%d,%d"
)   
```
